```python
import tensorflow as tf
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# plot examples

img1 = cv2.imread('cat_example.jpg',1)
img2 = cv2.imread('not_cat_example.jpg',1)

img1 = cv2.resize(img1, dsize=(128, 128))
img2 = cv2.resize(img2, dsize=(128, 128))

fig, (ax0, ax1) = plt.subplots(1, 2, figsize=(10,20))
ax0.imshow(cv2.cvtColor(img1, cv2.COLOR_BGR2RGB))
ax0.set_title('cat', fontsize=int(20))
ax0.axis('off')
ax1.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
ax1.set_title('not cat', fontsize=int(20))
ax1.axis('off')
plt.show(fig)
```

```python
# preprocess images

def preprocess_image(image):
    image = image / 255
    image_mean = np.array(image).reshape(128*128*3,1).mean(axis=0)[0]
    image_std = np.array(image).reshape(128*128*3,1).std(axis=0)[0]
    image = (image - image_mean)/image_std
    return image.reshape(-1,128,128,3).astype('float32')
    
img1_prep = preprocess_image(img1)
img2_prep = preprocess_image(img2)
```

```python
# load trained model

model_base = tf.keras.applications.MobileNetV2(weights=None, include_top=False ,input_shape=[*[128, 128], 3])

model = tf.keras.Sequential([model_base,
                            tf.keras.layers.GlobalAveragePooling2D(),
                            tf.keras.layers.Dense(1, activation='sigmoid')
                            ])

model.load_weights("main_model_weights.h5")
```

```python
# Converting a tf.Keras model to a TensorFlow Lite model

converter = tf.lite.TFLiteConverter.from_keras_model(model)
tflite_model = converter.convert()
tflite_model_name = "converted_main_model.tflite"
open(tflite_model_name, "wb").write(tflite_model)
```

```python
# Load TFLite model and allocate tensors
interpreter = tf.lite.Interpreter(model_path="converted_main_model.tflite")
interpreter.allocate_tensors()
# Get input and output tensors
input_details = interpreter.get_input_details()
output_details = interpreter.get_output_details()
```

```python
# Test TFLite model

def tflite_prediction(input_data):
    interpreter.set_tensor(input_details[0]['index'], input_data)
    interpreter.invoke()
    output_data = interpreter.get_tensor(output_details[0]['index'])[0][0]
    probability = '{:0.1f}'.format(100*output_data)
    return probability

cat_proba1 = tflite_prediction(img1_prep)
cat_proba2 = tflite_prediction(img2_prep)
```

```python
# plot images with probabilities

fig, (ax0, ax1) = plt.subplots(1, 2, figsize=(10,20))
ax0.imshow(cv2.cvtColor(img1, cv2.COLOR_BGR2RGB))
ax0.set_title(str(cat_proba1) + '% cat', fontsize=int(20))
ax0.axis('off')
ax1.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
ax1.set_title(str(cat_proba2) + '% cat', fontsize=int(20))
ax1.axis('off')
plt.show(fig)
```
